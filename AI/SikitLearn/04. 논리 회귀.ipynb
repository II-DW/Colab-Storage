# 4. Logisitc Regression

공부 시간에 따른 자격증 시험 합격 가능성

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LogisticRegressionData.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

데이터 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

###학습 (로지스틱 회귀 모델)

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

6시간 공부했을 때 예측?

```python
classifier.predict([[6]])
# 결과 1 : 합격할 것으로 예측
```

```python
classifier.predict_proba([[6]]) #합격할 확률 출력
# 불합격 확룰 14%, 합격할 확률 86%
```

4시간 공부했을 때 예측?

```python
classifier.predict([[4]])
#결과 0 : 불 합격할 것으로 예측
```

```python
classifier.predict_proba([[4]]) #합격할 확률 출력
# 불합격 확룰 63%, 합격할 확률 37%
```

### 분류 결과 예측 (테스트 세트)

```python
y_pred = classifier.predict(X_test)
y_pred # 예측 값
```

```python
y_test # 실제 값
```

```python
X_test # 공부 시간 정보
```

```python
classifier.score(X_test,y_test)
# 전체 테스트 세트 4개 중에서 분류 예측을 올바로 맞힌 개수 3개-> 4개 (75%)
```

### 데이터 시각화 (테스트 세트)

```python
X_range = np.arange(min(X), max(X), 0.1)
X_range
```

```python
p = 1 / (1 + np.exp(-(classifier.coef_ * X_range + classifier.intercept_))) # y = mx + b
p
```

```python
p.shape
```

```python
X_range.shape
```

```python
p = p.reshape(-1)
p.shape
```

> 시그모이드 함수

```python
plt.scatter(X_train , y_train , color = 'blue')
plt.plot(X_range, p, color = 'green')
plt.plot(X_range, np.full(len(X_range), 0.5), color = 'red') #X_range 개수만큼 0.5로 가득찬 배열 만들기
plt.title('Probalbilitiy by hours')
plt.xlabel('hours')
plt.ylabel('P')
plt.show()
```

### 데이터 시각화 (훈련 세트)

```python
plt.scatter(X_test , y_test , color = 'blue')
plt.plot(X_range, p, color = 'green')
plt.plot(X_range, np.full(len(X_range), 0.5), color = 'red') #X_range 개수만큼 0.5로 가득찬 배열 만들기
plt.title('Probalbilitiy by hours (test)')
plt.xlabel('hours')
plt.ylabel('P')
plt.show()
```
