```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x, y) :
  return 3 * x * x  + 3 * y * y + x * y * y + 7 * x + 7 * y + 10
```

```python
def g(x, y) :
  return 6 * x + y * y + 7
```

```python
def h(x, y) :
  return 6 * y + 2 * x * y + 7
```

```python
import numpy as np

fig = plt.figure(figsize=(9, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.arange(-4, 4, 0.1)
y = np.arange(-4, 4, 0.1)

xx, yy = np.meshgrid(x, y)
temp = np.zeros((len(x), len(y)))
for i in range (len(x)) :
  for j in range (len(y)) :
    temp[i, j] = f(x[i], y[j])

y_zeros = np.zeros(len(y))

ax.plot_surface(xx, yy, temp, rstride = 1, cstride = 1, alpha = 1, color = 'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

```python
import numpy as np

fig = plt.figure(figsize=(9, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.arange(-4, 4, 0.1)
y = np.arange(-4, 4, 0.1)

xx, yy = np.meshgrid(x, y)
temp = np.zeros((len(x), len(y)))
for i in range (len(x)) :
  for j in range (len(y)) :
    temp[i, j] = f(x[i], y[j])

y_zeros = np.zeros(len(y))
x_zeros = np.zeros(len(x))

ax.plot_surface(xx, yy, temp, rstride = 1, cstride = 1, alpha = 0.1, color = 'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.plot(x, y_zeros, f(x,0), color = 'red')
ax.plot(x, y_zeros, g(0,0) * x + f(0, 0), color = 'black')
ax.plot(x_zeros, y, f(0,x), color = 'red')
ax.plot(x_zeros, y, g(0,0) * y + f(0, 0), color = 'black')
ax.view_init(0, 0)
plt.show()
```

```python
def f(x, y) :
  return x * x + y * y - 0.5 * x * y + x - y
```

```python
def g(x, y) :
  return 2 * x - 2 * y + 1
```

```python
def h(x, y) :
  return 2 * y - 2 * x - 1
```

```python
import numpy as np

fig = plt.figure(figsize=(9, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)

xx, yy = np.meshgrid(x, y)
z = f(xx, yy)
surf = ax.plot_surface(xx,yy,z,rstride=2,cstride=2,cmap=plt.cm.coolwarm,linewidth=0.5,antialiased=True)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

plt.show()
```

```python

u = np.linspace(-4, 4, 100)
v = np.linspace(-4, 4, 100)
U, V = np.meshgrid(u, v)
Z = f(U, V)
# 경사하강법 시뮬레이션
W = np.array([4, -4.0])
W1 = [W[0]]
W2 = [W[1]]
N = 21
alpha = 0.05
for i in range(N):
    W = W - alpha *np.array([g(W[0], W[1]), h(W[0], W[1])])
    W1.append(W[0])
    W2.append(W[1])
```

```python
n_loop=20
```

```python
from matplotlib.animation import FuncAnimation
```

```python
# 그림 4-9 왼쪽
WW1 = np.array(W1[:n_loop])
WW2 = np.array(W2[:n_loop])
ZZ = f(WW1, WW2)
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')

def update (num, data, line) :
  line.set_data(data[:2, :num])
  line.set_3d_properties(data[2, :num])

ax.plot_surface(U, V, Z,cmap=plt.cm.coolwarm, alpha=0.6)
data = np.array([WW1, WW2, ZZ])
line, = ax.plot(data[0, 0:1], data[1, 0:1], data[2, 0:1], 'o-', c = 'k', alpha = 1, markersize = 7)

list_x = []
list_y = []


ani = FuncAnimation(fig, update, 20, fargs=(data, line), interval=1 , blit=False)
ani.save('matplot.gif', writer='imagemagick')

plt.show()
```

```python
# 그림 4-9 오른쪽
plt.figure(figsize=(8,8))
plt.contourf(U, V, Z, levels=[5,10,20,30,40,50,70,100, 200], cmap='Blues')
C = plt.contour(U, V, Z, levels=[5,10,20,30,40,50,70,100, 200],colors='k')
plt.clabel(C, inline=1, fontsize=10, fmt='%r')
plt.plot(W1[:n_loop], W2[:n_loop], '-o', c='k')
plt.gca().set_aspect('equal')
plt.xticks(range(-4,4,1))
plt.yticks(range(-4,4,1))
plt.xlabel('u', fontsize=14)
plt.ylabel('v', fontsize=14)
plt.grid(linewidth=2)
plt.show()
```
