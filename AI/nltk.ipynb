```python
import pandas as pd
df = pd.read_csv('급식식단정보.csv')

MenuList = []
for i in range (0, len (df)) :
  MenuOneDayList=[]
  for s in df['요리명'][i].split('<br/>') :
    FinalSTR = ''
    for f in s :
      if f == ' ' or f == '(': # 문자 제외 나머지 데이터 삭제
        break
      FinalSTR+=f
    MenuOneDayList.append(FinalSTR)
  MenuList.append(MenuOneDayList)
MenuList
```

```python
from nltk.util import ngrams
```

```python
from nltk.lm.preprocessing import pad_both_ends
```

```python
sentences = []
for d in MenuList:
  bigram = ngrams(d, 2, pad_left=True, pad_right=True, left_pad_symbol="", right_pad_symbol="")
  sentences += [t for t in bigram]
sentences
```

```python
from nltk.lm.preprocessing import padded_everygram_pipeline
train, vocab = padded_everygram_pipeline(2, sentences)
```

```python
from nltk.lm import MLE
lm = MLE(2) # bigram
lm.fit(train, vocab)
```

```python
print(lm.generate(7, text_seed =['고구마조림']))
```
