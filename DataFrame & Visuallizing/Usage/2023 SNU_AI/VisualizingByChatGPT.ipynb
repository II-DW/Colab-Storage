```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Convert the data and target into a DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Create a scatter plot to visualize the data
plt.figure(figsize=(10, 6))

# Plot each class with a different color
for target_class in iris.target_names:
    class_data = iris_df[iris_df['target'] == iris.target_names.tolist().index(target_class)]
    plt.scatter(class_data['sepal length (cm)'], class_data['sepal width (cm)'], label=target_class)

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset Visualization')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Convert the data and target into a DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['target_name'] = [iris.target_names[i] for i in iris.target]
iris_df = iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)','target_name']]

# Calculate the average values grouped by target
averages = iris_df.groupby('target_name').mean()

# Create a bar graph
averages.plot(kind='bar', figsize=(10, 6))
plt.title('Average Feature Values by Target')
plt.ylabel('Average Value')
plt.xlabel('Target')
plt.xticks(rotation=0)
plt.legend(title='Target Name')
plt.show()
```
