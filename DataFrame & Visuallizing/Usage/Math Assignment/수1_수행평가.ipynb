```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.linspace(-8 * np.pi, 8 * np.pi, 2000)
Y=np.sin(X)
```

```python
plt.plot(X, Y)
plt.axvline(x=0, color = 'black')  # draw x =0 axes
plt.axhline(y=0, color = 'black')   # draw y =0 axes
plt.title('y = sin Î¸')
plt.show()
```

```python
Y2=np.cos(X)
plt.plot(X, Y2)
plt.axvline(x=0, color = 'black')  # draw x =0 axes
plt.axhline(y=0, color = 'black')   # draw y =0 axes
plt.title('y = cos Î¸')
plt.show()
```

```python
X=np.linspace(0, 128, 200000)
Y1 = 15 * np.sin(2 * np.pi * X)
Y2 = 15 * np.sin(4 * np.pi * X)
Y3 = 15 * np.sin(6 * np.pi * X)
Y4 = 15 * np.sin(8 * np.pi * X)
Y5 = 15 * np.sin(10 * np.pi * X)
```

```python
sub_plots = plt.subplots(5, 1, figsize=(10,50))
fig = sub_plots[0]
graph = sub_plots[1]

graph[0].plot(X, Y1)
graph[1].plot(X, Y2)
graph[2].plot(X, Y3)
graph[3].plot(X, Y4)
graph[4].plot(X, Y5)

graph[0].set_title('15 * sin 2t')
graph[1].set_title('15 * sin 4t')
graph[2].set_title('15 * sin 6t')
graph[3].set_title('15 * sin 8t')
graph[4].set_title('15 * sin 10t')


graph[0].axvline(x=0, color = 'black')  # draw x =0 axes
graph[0].axhline(y=0, color = 'black')   # draw y =0 axes

graph[1].axvline(x=0, color = 'black')  # draw x =0 axes
graph[1].axhline(y=0, color = 'black')   # draw y =0 axes

graph[2].axvline(x=0, color = 'black')  # draw x =0 axes
graph[2].axhline(y=0, color = 'black')   # draw y =0 axes

graph[3].axvline(x=0, color = 'black')  # draw x =0 axes
graph[3].axhline(y=0, color = 'black')   # draw y =0 axes

graph[4].axvline(x=0, color = 'black')  # draw x =0 axes
graph[4].axhline(y=0, color = 'black')   # draw y =0 axes

graph[0].set_xlim([-0.1, 2])
graph[1].set_xlim([-0.1, 2])
graph[2].set_xlim([-0.1, 2])
graph[3].set_xlim([-0.1, 2])
graph[4].set_xlim([-0.1, 2])

plt.show()
```
