```python
def calculateCrossProduct(Vector1, Vector2) :
  return (Vector1[1] * Vector2[2]  - Vector1[2] *Vector2[1],
          Vector1[2] * Vector2[0]  - Vector1[0] *Vector2[2],
          Vector1[0] * Vector2[1]  - Vector1[1] *Vector2[0] )
```

```python
def CwCcw(Vector1, Vector2) :
  return   Vector1[0] * Vector2[1]  - Vector1[1] *Vector2[0]
```

```python
In_Processing_dots = [Start_dot, dot_list_remain[0][:2]] # First, Second
dot_list_remain = dot_list_remain[1:]
```
