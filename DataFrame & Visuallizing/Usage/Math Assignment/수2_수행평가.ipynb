```python
def MSEw (x, y, w,b) :
  return (2/len(x)) * sum((-y + w * x + b) * x)

def MSEb (x, y, w,b) :
  return (2/len(x)) * sum((-y + w * x + b))
```

```python
import numpy as np

def gradient(x_data, y_data, w, b,lr) :
  x_data = np.array(x_data)
  y_data = np.array(y_data)

  #현재 오차 그래프에서 w의 미분계수
  w_gradient = MSEw(x_data, y_data, w, b)

  #현재 오차 그래프에서 b의 미분계수
  b_gradient = MSEb(x_data, y_data, w, b)

  w = w - lr * w_gradient
  b = b - lr * b_gradient

  return w, b
```

```python
epochs = 6001  #학습 횟수
lr = 0.00000025   #학습률
w = 0       #최초 기울기 설정
b = 0       #최초 y절편 설정

List = []

#학습 횟수 epoch만큼 경사 하강법 실행
for i in range(epochs):
  w, b = gradient(x, y, w, b, lr)

  # 학습 횟수가 1000번째일 때마다 기울기와 y절편을 출력
  if i < 40 :
    print('학습 횟수', i, '기울기', w, 'y절편', b)
    List.append([w, b])
```

```python
from matplotlib.animation import FuncAnimation

fig = plt.figure()
ax = plt.axes()

X = np.array(x)
Y = List[0][0] * X + List[0][1]
line, = ax.plot(X, Y, color='red')

def update(frame):
    Y=List[frame][0] * X + List[frame][1]

    line.set_data(X, Y)
    return line,

plt.scatter(x, y)

ani = FuncAnimation(fig, update, len(List), interval=1 , blit=False)
ani.save('matplot.gif', writer='imagemagick')

plt.show()
```
