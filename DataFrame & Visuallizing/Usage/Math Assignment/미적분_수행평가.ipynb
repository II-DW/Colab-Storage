```python
import math
def f(x, limit_n) :
  return_value = 0
  for n in range (1, limit_n+1) :
    if n % 4 == 1 :
      return_value += (x**n) / math.factorial(n)
    elif n % 4 == 3 :
      return_value -= (x**n) / math.factorial(n)
  return return_value
```

```python
import numpy as np
import matplotlib.pyplot as plt

X=np.linspace(0, 4, 2000)
n=1
Y=f(X, n)

Y2 = np.sin(X)

plt.plot(X, Y)
plt.plot(X, Y2, color = 'red')

plt.axvline(x=0, color = 'black')  # draw x =0 axes
plt.axhline(y=0, color = 'black')   # draw y =0 axes
plt.title(str(n) + ' done')
plt.show()
```
