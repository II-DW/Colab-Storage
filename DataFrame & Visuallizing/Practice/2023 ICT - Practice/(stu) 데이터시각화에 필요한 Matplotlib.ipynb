```python
import pandas as pd
import numpy as np
# 맷플롯립 중 그래프를 그리는 pyplot 모듈을 불러와 plt라는 별명으로 사용
from matplotlib import pyplot as plt
```

```python
# 한글 해결 테스트
plt.rc('font', family='NanumBarunGothic') #폰트 설정
plt.title("한글 제목 출력 테스트") # 그래프 제목 출력
# plt.plot()
# plt.show()
```

```python
# 선그래프 그리기
# plt.plot([x값],[y값])
# plt.show()

plt.plot([1, 4, 9, 16, 25]) #데이터가 하나 있으면 y축!
plt.show()
```

```python
# 선그래프 그리기
# x의 범위가 1,2,3,4,5이면서 y = x제곱 관계를 나타내는 선그래프

plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) #(x축, y축)
plt.show()

# x = np.arange(1, 6) # 1,2,3,4,5
# y = x**2
# plt.plot(x, y)
# plt.show()
```

```python
# 산점도 그래프 그리기 > 데이터간의 관계를 좌표상의 점으로 시각화

x = np.arange(1, 5)
y = np.arange(2, 6)
plt.scatter(x, y)
plt.show( )
```

```python
# 막대그래프 그리기

x = ['mon', 'tue', 'wed', 'thu', 'fri']
y = [50, 40, 20, 60, 35]
plt.bar(x, y)
plt.show( )
```

```python
# 파이차트 그리기 > 범주형 데이터의 비율을 원형으로 표시
# plt.pie(영역별 비율,labels=영역별 레이블,autopct='표시할 숫자 형식')

ratio = [20, 30, 40, 60]
label = ['Amy', 'Betty', 'Chris', 'Dorothy']
plt.pie(ratio, labels = label,autopct = '%1.1f%%') #범례 표시, 데이터 값 표시
plt.show()
```

```python
# 히스토그램 그리기 > 수치형 데이터를 동일한 폭으로 그룹화하여 시각화
# plt.hist(데이터 리스트, bins=계급 개수),  계급 개수: 기본값은 10

data = [3, 4, 5, 5, 5, 6, 7, 8, 3,
        2, 7, 7, 7, 8, 9, 10, 10, 9]
plt.hist(data, bins = 5, color = 'b') # 히스토그램
plt.show()
```

```python
#히트맵 그리기
#plt.matshow(배열명)

x = np.arange(1, 5)
y = np.arange(2, 6)
arr = ((x, y))
plt.matshow(arr)
plt.show()
```

```python
# 마커, 격자무늬 표현하기

x = np.arange(1, 5)
y = np.arange(2, 6)
plt.scatter(x, y, marker = 'D') # 마커 표시
plt.grid() # 격자무늬 표현
plt.show( )
```

```python
# 그래프 색상 표현하기
# plt.그래프명(···, color='색상 기호')

x = np.arange(1, 5)
y = np.arange(2, 6)
plt.plot(x, y,linestyle = '--', color = 'y',marker = 'D') # 색상 표현
plt.show( )
```

```python
# 축 레이블 표시
x = np.arange(1, 5)
y = np.arange(2, 6)

plt.plot(x, y)
plt.xlabel('X', fontsize = 14, labelpad = 20) # x축 제목 표시
plt.ylabel('Y', fontsize = 20, labelpad = 10) # y축 제목 표시
plt.show( )
```

```python
# 범례 표시하기
# 범례 위치 지정 : upper left, upper right, lower left, lower right
x = np.arange(1, 5)
y = x ** 2

plt.plot(x, y, label = 'y = x^2')
plt.legend(loc = 'upper left') # 범례 표시
plt.show()
```

```python
# 축 범위 지정하기, 축 눈금 지정하기

x = np.arange(1, 5)
y = x ** 2

plt.plot(x, y)
plt.xlim([0, 5]) # x축 범위 지정
plt.ylim([0, 20]) # y축 범위 지정
#plt.axis([0, 5, 0, 20])  # x축, y축 범위 지정

#plt.xticks(np.arange(0,6)) # x축 눈금 지정
#plt.xticks(np.arange(0,6),('10', '20', '30', '40', '45', 'TOP')) # x축 눈금 지정
#plt.yticks(np.arange(0,21,5)) # y축 눈금 지정
plt.show( )
```

```python
# 그래프 여러 개 한꺼번에 그리기

x = np.arange(1, 5)

plt.plot(x, x * 2, marker = 'o', color = 'y', label = 'y = x * 2')
plt.plot(x, x + 3, marker = 's', color = 'c', label = 'y = x + 3')
plt.plot(x, x ** 2, marker = '*', color = 'm', label = 'y = x ** 2')
plt.legend()
plt.show( )
```

#### (연습문제)
* 고양이 가족 체중 그래프를 그리시오.

```python
import pandas as pd
import matplotlib.pyplot as plt

family = {'age': [60, 70, 12, 45, 42],
           'weight': [3.8, 7.2, 6.3, 5.5, 4.6]}

cat = pd.DataFrame(family, ['grandpa', 'grandma', 'grandson', 'papa', 'mama'])
cat
```

```python
# 고양이 가족 체중 그래프
x = cat.index
plt.plot(cat.index, cat['weight'], color = 'm', linestyle = '--',marker = '*')
plt.xlabel('family', fontsize = 14, labelpad = 20)
plt.ylabel('weight', fontsize = 14, labelpad = 10)
plt.title('고양이 가족 체중 그래프')
#
#
#
plt.show()
```
