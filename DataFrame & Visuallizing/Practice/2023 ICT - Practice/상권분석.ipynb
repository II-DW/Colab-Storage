```python
df = df[['상가업소번호', '상호명', '상권업종대분류명',
       '상권업종중분류명', '상권업종소분류명', '시군구명',  '도로명주소',
        '경도', '위도']]
```

```python
df
```

```python
df['상권업종대분류명'].unique()
```

```python
y = []
x = df[df['상권업종대분류명']=='관광/여가/오락']['상권업종중분류명'].unique()
for i in x :
  y.append(len(df[df['상권업종중분류명']==i]))
```

```python
colorList = []
for n in range (len(y)) :
  if n == 1 :
    colorList.append('red')
    continue
  colorList.append('blue')
```

```python
plt.bar(x, y, color=colorList)
plt.xticks(rotation=60)

for i, x in enumerate(x) :
  c = 'blue'
  if i == 1 :
    c = 'red'
  plt.text(x, y[i], y[i],
           fontsize=9,
           color=c,
           horizontalalignment='center',
           verticalalignment='bottom'
           )
plt.show()
```

```python
y = []
x = df[df['상권업종대분류명']=='학문/교육']['상권업종중분류명'].unique()
for i in x :
  y.append(len(df[df['상권업종중분류명']==i]))
```

```python
colorList = []
for n in range (len(y)) :
  if n == 8 or n == 5 :
    colorList.append('red')
    continue
  colorList.append('blue')
```

```python
plt.bar(x, y, color=colorList)
plt.xticks(rotation=60)

for i, x in enumerate(x) :
  c = 'blue'
  if i == 8 or i == 5:
    c = 'red'
  plt.text(x, y[i], y[i],
           fontsize=9,
           color=c,
           horizontalalignment='center',
           verticalalignment='bottom'
           )
plt.show()
```

```python
import folium
```

```python
map=folium.Map(location=[35.845115, 128.627126], zoom_start=14) #창융원 중심
map
```

```python
# 지도에 마커 표시하기
#반복문으로 마킹하기
for i in df[df['상권업종중분류명']=='학원-컴퓨터'].index:
  name = df.loc[i, '상호명']
  lat = df.loc[i, '위도']
  lon = df.loc[i, '경도']
  marker = folium.Marker([lat,lon], tooltip=name).add_to(map)
map
```
