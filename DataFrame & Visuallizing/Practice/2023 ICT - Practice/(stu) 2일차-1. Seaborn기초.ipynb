# Seaborn

Seaborn은 데이터 시각화를 위한 파이썬 라이브러리로, Matplotlib를 기반으로 만들어졌으며 보다 간편하고 멋진 그래프를 그릴 수 있도록 도와줍니다. Seaborn의 주요 함수들과 기능에 대해 알려드리겠습니다:
- sns.scatterplot(): 산점도를 그릴 때 사용하며, 두 변수 간의 관계를 시각화합니다.
- sns.lineplot(): 선 그래프를 그릴 때 사용하며, 시간 또는 연속적인 값에 대한 변화를 보여줍니다.
- sns.barplot(): 막대 그래프를 그릴 때 사용하며, 범주별 값의 크기를 비교합니다.
- sns.countplot(): 막대 그래프의 특수한 형태로, 각 범주의 빈도를 표시합니다.
- sns.boxplot(): 상자 그림을 그릴 때 사용하며, 데이터의 분포와 이상치를 시각화합니다.
- sns.violinplot(): 바이올린 플롯을 그릴 때 사용하며, 밀도 추정과 상자 그림을 함께 보여줍니다.
- sns.heatmap(): 히트맵을 그릴 때 사용하며, 두 변수 간의 상관관계 또는 행렬 데이터를 시각화합니다.
- sns.pairplot(): 산점도 행렬을 그릴 때 사용하며, 데이터프레임의 수치형 변수들 간의 모든 조합에 대한 산점도를 한 번에 그립니다.
- sns.kdeplot(): 밀도 추정 그래프를 그릴 때 사용하며, 커널 밀도 추정을 통해 연속적인 분포를 시각화합니다.
- sns.jointplot(): 두 변수의 산점도와 히스토그램을 함께 그릴 때 사용합니다.
- sns.distplot(): 히스토그램과 커널 밀도 추정을 함께 그릴 때 사용하며, 분포를 시각화합니다.
- sns.boxenplot(): 상자 그림의 확장 버전으로, 데이터의 분포를 더 자세하게 시각화합니다.
- sns.swarmplot(): 스웜 플롯을 그릴 때 사용하며, 데이터의 분포를 개별 데이터 포인트로 보여줍니다.
- sns.lmplot(): 선형 회귀 선과 산점도를 함께 그릴 때 사용하며, 두 변수 간의 선형 관계를 시각화합니다.
- sns.regplot(): 선형 회귀 선과 산점도를 함께 그릴 때 사용하며, 두 변수 간의 선형 관계를 시각화합니다.

# Matplotlib의 서브플롯(Subplot) 기능
- 여러개의 그래프를 한꺼번에 나타내기

```python
# 샘플 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.exp(x)

# 서브플롯 설정 (2행 2열로 그래프 배치)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 첫 번째 그래프: 선 그래프
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('선 그래프')

# 두 번째 그래프: 산점도
axes[0, 1].scatter(x, y2)
axes[0, 1].set_title('산점도')

# 세 번째 그래프: 막대 그래프
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 25, 10]
axes[1, 0].bar(categories, values)
axes[1, 0].set_title('막대 그래프')

# 네 번째 그래프: 히스토그램
data = np.random.randn(100)
axes[1, 1].hist(data)
axes[1, 1].set_title('히스토그램')

# 서브플롯 간 간격 조절
plt.tight_layout()

# 전체 그림에 제목 추가
plt.suptitle('다양한 그래프 예제')

# 그래프 표시
plt.show()
```

1. 관계 그래프 (Pair Plot):


```python
import seaborn as sns
iris = sns.load_dataset('iris')

# 관계 그래프
sns.pairplot(iris, hue='species') # hue: 카테고리 변수
```

2. 박스 플롯 (Box Plot):

```python
sns.boxplot(x='species', y='sepal_length', data=iris, color='orange')
sns.boxplot(x='species', y='sepal_width', data=iris, color='purple')
sns.boxplot(x='species', y='petal_length', data=iris, color='green')
sns.boxplot(x='species', y='petal_width', data=iris, color='skyblue')
# sns.boxplot(x='species', y='sepal_length', data=iris, palette='Set1')

plt.show()
```

3.  밀도 추정 그래프 (KDE Plot):


```python
sns.kdeplot(data=iris, x='sepal_length', hue='species')
sns.kdeplot(data=iris, x='sepal_width', hue='species')
sns.kdeplot(data=iris, x='petal_length', hue='species')
sns.kdeplot(data=iris, x='petal_width', hue='species')
plt.show()
```

4. 밀집도 플롯 (Violin Plot):


```python
sns.violinplot(x='species', y='sepal_length', data=iris)
sns.violinplot(x='species', y='sepal_width', data=iris)
sns.violinplot(x='species', y='petal_length', data=iris)
sns.violinplot(x='species', y='petal_width', data=iris)
plt.show()
```

# 데이터 셋 (Dataset) 다루기

`sklearn.dataset` 안에는 빌트인 (built-in) 데이터 셋들이 존재합니다. 물론 튜토리얼 진행을 위한 수준이므로, 규모가 크지는 않습니다 (Toy Dataset 이라고도 불리웁니다.)

그렇지만, mldata.org 데이터 셋은 조금 더 규모가 큰 데이터 셋을 제공하며, 온라인에서 동적으로 다운로드를 받을 수 있습니다.

이번 튜토리얼에서는 Built-in 데이터 셋을 활용하는 방법에 대해서 알아보도록 하겠습니다.

## 빌트인 (Built-in) 데이터셋 활용

- `load_boston`: 보스톤 집값 데이터
- `load_iris`: 아이리스 붓꽃 데이터
- `load_diabetes`: 당뇨병 환자 데이터
- `load_digits`: 손글씨 데이터
- `load_linnerud`: multi-output regression 용 데이터
- `load_wine`: 와인 데이터
- `load_breast_cancer`: 위스콘신 유방암 환자 데이터

### 데이터 셋 조회

빌트인 데이터셋은 `sklearn.utils.Bunch` 라는 자료구조를 활용합니다.

**key-value** 형식으로 구성되어 있으며, 사전(dict)형 타입과 유사한 구조를 가지고 있습니다.

공통 **key**는 다음과 같습니다.

- `data`: 샘플 데이터, Numpy 배열로 이루어져 있습니다.
- `target`: Label 데이터, Numpy 배열로 이루어져 있습니다.
- `feature_names`: Feature 데이터의 이름
- `target_names`: Label 데이터의 이름
- `DESCR`: 데이터 셋의 설명
- `filename`: 데이터 셋의 파일 저장 위치 (csv)

3.   데이터의 특성 사이의 상관관계를 분석해 보세요.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# 와인 데이터셋 불러오기
wine = load_wine()

# 데이터셋을 DataFrame으로 변환
# np.c_[] 함수를 사용하여 입력 데이터와 타겟 데이터를 열 방향으로 합침
wine_df = pd.DataFrame(data=np.c_[wine['data'], wine['target']],
                       columns=wine['feature_names'] + ['target'])
```

와인 데이터의 특징 설명:

레이블 분포 확인: 데이터셋에는 세 가지 종류의 와인이 있으며, 각 클래스별로 데이터가 균형있게 분포되어 있습니다.

특성 분포 확인: 특성들 간에 범위가 다르므로 스케일링을 고려해야 합니다.

특성 간 상관관계: 'flavanoids'와 'total_phenols' 사이에 강한 양의 상관관계가 있음을 확인할 수 있습니다. 그 외에도 다른 특성들 사이에 상관관계가 보일 수 있습니다.

'alcohol' 특성과 레이블의 관계: 와인의 종류별로 알코올 함유량에 차이가 있는 것으로 보입니다.

'flavanoids' 특성과 레이블의 관계: 플라보노이드 함유량이 높을수록 특정 와인 종류에 속할 가능성이 높은 것으로 보입니다.

```python
wine_df
```

```python
# 1. 레이블 분포 확인 (카운트 플롯)
sns.countplot(x='target', data=wine_df)
plt.xlabel('와인의 종류')
plt.ylabel('데이터 개수')
plt.title('와인 데이터 레이블 분포')
plt.show()
```

```python
# 2. 특성 분포 확인 (박스 플롯)
sns.boxplot(data=wine_df.drop('target', axis=1))
plt.xticks(rotation=45)
plt.title('와인 특성 분포')
plt.show()
```

```python
# 3. 특성 간 상관관계 (히트맵)
correlation_matrix = wine_df.corr()
plt.figure(figsize=(15, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('와인 특성 간 상관관계')
plt.show()
```

```python
# 4. 특정 특성과 레이블의 관계 (박스 플롯)
sns.boxplot(x='target', y='alcohol', data=wine_df)
plt.title('와인 종류별 알코올 함유량')
plt.xlabel('와인의 종류')
plt.ylabel('알코올 함유량')
plt.show()
```

```python
# 5. 특정 특성과 레이블의 관계 (바이올린 플롯)
sns.violinplot(x='target', y='flavanoids', data=wine_df)
plt.title('와인 종류별 플라보노이드 함유량')
plt.xlabel('와인의 종류')
plt.ylabel('플라보노이드 함유량')
plt.show()
```
