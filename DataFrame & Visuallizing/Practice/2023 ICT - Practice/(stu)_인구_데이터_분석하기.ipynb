```python
# 관련 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 업로드하기 #encoding='cp949
df=pd.read_csv('/content/202306_202306_연령별인구현황_월간.csv', encoding='cp949', thousands=',', index_col='행정구역')
```

```python
# 데이터 확인하기
df
```

```python
dong = input('원하는 지역의 이름을 입력해주세요 : ')  # 동 이름 입력
# 3 동이 포함된 행을 찾아서 인구 수를 저장
df_area1 = df[df.index.str.contains(dong)]
```

```python
# df_area1을 숫자형으로 변환 문자->숫자로 변경 안되는것은 NaN으로 변경
df_area1
```

```python
# 행열 전환하기
df_area1 = df_area1.T
```

```python
df_area1
```

```python
df_area1 = df_area1.drop(df_area1.index[0])
df_area1 = df_area1.drop(df_area1.index[0])
df_area1
```

```python
#인구분포 차트 그리기
import matplotlib.pyplot as plt
List = [i for i in range(1, 102)]
plt.bar(List, df_area1['대구광역시 중구 동인동(2711051700)'])
```

```python
df_area2 = df_area1.reset_index()[['대구광역시 중구 동인동(2711051700)']]
df_area2
```

```python
#데이터 분석 범위를 10대로 한정하기
df_area2[10:20]
```

```python
# 2개 동을 선택하고 2개동의 인구 데이터 비교
dong1, dong2 = input('원하는 지역의 이름을 입력해주세요 : ').split()  # 동 이름 입력
# 3 동이 포함된 행을 찾아서 인구 수를 저장
df_area1 = df[df.index.str.contains(dong1)]
df_area2 = df[df.index.str.contains(dong2)]
```

```python
type(df_area1)
```

```python
# 동 전체의 인구를 분석해보자
# ex. 평리 1,2,3,...동 데이터를 묶어 평리동으로 만들자!
df_new = pd.concat([df_area1, df_area2]).T
```

```python
# 불필요한 열 삭제해보기
df_new = df_new.drop(df_new.index[0])
df_new = df_new.drop(df_new.index[0])
df_new
```
